```python
import renom as rm 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(-6, 6, 0.1)
f = rm.relu(x)
g = rm.swish(x)
```

```python
plt.title("ReLU and Swish functions", fontsize = 16)
plt.plot(x, f, label='ReLU', linewidth = 3)
plt.plot(x, g, label= 'Swish', linewidth = 3)
plt.legend(prop={'size': 14})
plt.grid()
plt.axes()
plt.show()
```

```python
x = np.arange(-7, 7, 0.1)
f = (x > 0) * 1
g = x / (1. + np.exp(-x)) + (1. / (1. + np.exp(-x))) * (1. - x * (1. / (1. + np.exp(-x))))
```

```python
plt.title("ReLU and Swish function 1st derivatives", fontsize = 16)
plt.plot(x, f, label='dReLU', linewidth = 3)
plt.plot(x, g, label='dSwish', linewidth = 3)
plt.legend(prop={'size': 14})
plt.grid()
plt.axes()
plt.show()
```
